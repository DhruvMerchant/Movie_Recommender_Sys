```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get('https://www.imdb.com/list/ls074451163/').text
 
```

```python
bsoup = BeautifulSoup(response, 'lxml')
bsoup
```

```python
data = bsoup.findAll('div',class_='lister-item mode-detail') 
```

```python
name = []
rating = []
yr =[]
duration = []
votes = []
gross = []
genres = []
```

```python
for i in data:
    temp = i.h3.a.text
    name.append(temp)
    
    temp2 = i.h3.find('span', class_='lister-item-year text-muted unbold').text.strip('()')
    yr.append(temp2)
    
    temp3 = i.find('span', class_='ipl-rating-star__rating').text
#     print(temp3)
    rating.append(temp3)
    
    temp4 = i.p.find('span', class_='runtime').text
#     print(temp4)
    duration.append(temp4)
    
    temp5= i.findAll('span',attrs={'name':'nv'})
    
    votes.append(temp5[0].text)
    gross.append((temp5[1].text) if len(temp5)>1 else "N.A")
#     print(gross)
    
    temp6 = i.find('span', class_='genre').text.split(',')
    temp6 = [x.strip() for x in temp6]
    genres.append(temp6)
    print(temp6)
```

```python
df = pd.DataFrame({'title': name, 'rating':rating,'release_year':yr,'gross_margin':gross,'genres':genres,'votes':votes,'runtime':duration})
```
